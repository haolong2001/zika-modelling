```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sigma2 = 0.264  # Standard deviation of noise
days = np.arange(1, 1000)  # Day 1 to 1000

# Calculate Rt for each day
Rt_values = []
Rt = 0.8
for day in days:
    seasonal_variation = 0.072 * np.sin(2 * np.pi * day / 365 - 0.18)
    noise = np.random.normal(0, sigma2)
    Rt = 1 + seasonal_variation + noise
    Rt = max(0.2, Rt)  # Ensure Rt is larger than 0.2
    Rt = min(Rt, 1.8)  # Ensure Rt is smaller than 1.8
    Rt_values.append(Rt)

# Plot Rt over days
plt.figure(figsize=(10, 6))
plt.plot(days, Rt_values, label='Rt', color='b')
plt.axhline(y=1, color='r', linestyle='--', label='Rt = 1 threshold')
plt.ylim(-1, 2)
# Labels and title
plt.xlabel('Days')
plt.ylabel('Rt (Reproduction Number)')
plt.title('ZIKV Rt Over Time with Seasonal Variation and Noise')
plt.legend()
plt.grid(True)


# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter1d

# Parameters
sigma2 = 0.264  # Standard deviation of noise
days = np.arange(1, 1000)  # Day 1 to 1000

# Calculate Rt for each day
Rt_values = []
Rt = 0.8
for day in days:
    seasonal_variation = 0.072 * np.sin(2 * np.pi * day / 365 - 0.18)
    noise = np.random.normal(0, sigma2)
    ###Rt = R_t=R_{t-1}+S_t+ N_t
    Rt = Rt   + noise  + seasonal_variation
    Rt = max(0.2, Rt)  # Ensure Rt is larger than 0.2
    Rt = min(Rt, 1.8)  # Ensure Rt is smaller than 1.8
    Rt_values.append(Rt)

# Apply 7-day moving average
Rt_smoothed = uniform_filter1d(Rt_values, size=7)

# Plot the smoothed Rt values
plt.figure(figsize=(10, 6))
plt.plot(days, Rt_smoothed, label='7-day Moving Average of Rt', color='b')
plt.axhline(y=1, color='r', linestyle='--', label='Rt = 1 threshold')

# Labels and title
plt.xlabel('Days')
plt.ylabel('Smoothed Rt (Reproduction Number)')
plt.title('ZIKV Rt (7-day Moving Average) Over Time with Seasonal Variation and Noise')
plt.ylim(-1, 2.5)  # Set y-axis limits
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
